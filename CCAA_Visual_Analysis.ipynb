# CCAA Visual Analysis from 2015 - 2019

```python
# The CCAA tournament is set up for 6 teams / seeded tournament.  
# These visuals are estimates based on CCAA data.
# An example situation would be if a team is looking for guaranteed spot in the playoff tournament.  
```

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
from sklearn import linear_model 
sns.set(rc={'figure.figsize':(10,8)})
sns.set_style('white')
```

```python
df = pd.read_csv('CCAA_.csv')
df.head()
```

```python
df['RD'] = df['RS'] - df['RA']
```

```python
df = df[df.Year <= 2019]
```

```python
no_tournament = df[df.Playoffs == 0]
yes_tournament = df[df.Playoffs == 1]
```

```python
# Conf_W & Runs Scored
plt.figure(figsize=(10,8))

plt.scatter(x=no_tournament.Conf_W,y = no_tournament.RS,c="red")

plt.scatter(x=yes_tournament.Conf_W,y = yes_tournament.RS,c="blue")

plt.xlabel("Conference Wins")

plt.ylabel("Runs Scored")

plt.axvline(x = 25)
```

```python
# Conf_W and Run Differential
plt.figure(figsize=(10,8))

plt.scatter(x=no_tournament.Conf_W,y = no_tournament.RD,c="red")

plt.scatter(x=yes_tournament.Conf_W,y = yes_tournament.RD,c="blue")

plt.xlabel("Conference Wins")

plt.ylabel("Runs Differential")

plt.axvline(x = 25)
```

```python
# Run Differential and Conf Wins 
plt.figure(figsize=(10,8))

plt.scatter(x=no_tournament.RD,y = no_tournament.Conf_W,c="red")

plt.scatter(x=yes_tournament.RD,y = yes_tournament.Conf_W,c="blue")

plt.xlabel("Run Differential")

plt.ylabel("Conference Wins")

plt.axvline(x = 100)
```

```python
# Run Differential and Conf Winning Percentage 
plt.figure(figsize=(10,8))

plt.scatter(x=no_tournament.RD,y = no_tournament.Conf_Pct,c="red")

plt.scatter(x=yes_tournament.RD,y = yes_tournament.Conf_Pct,c="blue")

plt.xlabel("Run Differential")

plt.ylabel("Conference Winning Percentage")

plt.axvline(x = 100)
```

# Regression Slope

```python
# Slope of Conference Wins and RD
x = np.array(df.RD)

y = np.array(df.Conf_W)

slope, intercept = np.polyfit(x, y, 1)

abline_values = [slope * i + intercept for i in x]

plt.figure(figsize=(10,8))

plt.scatter(x, y)

plt.plot(x, abline_values, 'r')

plt.title("Slope = %s" % (slope), fontsize = 12)

plt.xlabel("Run Differential " , fontsize =20)

plt.ylabel("Conf_Wins", fontsize = 20)

plt.axhline(25, 0, 1, color = "k", ls = '--')

plt.show()
```

```python
corrcheck = df[['RD', 'Conf_W', 'Playoffs']].copy()
g = sns.pairplot(corrcheck, hue = 'Playoffs', vars=["RD", "Conf_W"])
g.fig.set_set_size_inches(12,8)
```
