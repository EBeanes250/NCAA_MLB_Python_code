```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

# Walker Buehler

```python
df = pd.read_csv('Buehler.csv')
```

```python
df.head(20)
```

```python
df.sample(5)
```

```python
fig = plt.figure(figsize=(10,10))

fig = sns.scatterplot(data =df, x = 'plate_x', y = 'plate_z',hue= 'pitch_name', 
                      s = 80, edgecolor = 'black')

fig.set_aspect('equal')

sns.despine(left=True, bottom=True)

sz = plt.Rectangle(((-(17/2)/12),1.5), 17/12, 2,fill=None, ec="black", 
                   zorder=3,alpha=.7,lw=5)

plt.gca().add_patch(sz)

plt.xlabel('')

plt.ylabel('')
```

```python
df.columns
```

```python
strikes =['swinging_strike', 'called_strike','swinging_strike_blocked']
df['is_strike']=0
df.loc[df['description'].isin(strikes),'is_strike']=1

fig = plt.figure(figsize=(10,10))
fig = sns.kdeplot(data=df.loc[df['is_strike']==1], x='plate_x', y='plate_z', fill=True, color='red')

fig.set_aspect('equal')
sz = plt.Rectangle(((-(17/2)/12),1.5), 17/12, 2,fill=None, ec="black",zorder=3,alpha=.7,lw=5)
plt.gca().add_patch(sz)

sns.despine(left=True, bottom=True)
plt.xlabel('')
plt.ylabel('')
```
